```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Check the data
df_orders = pd.read_csv('../data/orders.csv')
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'])
df_orders.info()
```

```python
df_orders.head(5)
```

```python
df_orders = df_orders.sort_values(by=['customer_id', 'order_date'])
```

```python
gaps = df_orders.groupby('customer_id')['order_date'].diff().dt.days
```

```python
# Check: Only customers with 2+ orders
customers_order_count = df_orders.groupby('customer_id').size()
customers_with_multiple = customers_order_count[customers_order_count > 1]
print(f"Total customers: {len(customers_order_count)}")
print(f"Customers with 2+ orders: {len(customers_with_multiple)}")
print(f"Customers with 1 order: {len(customers_order_count[customers_order_count == 1])}")

# Only calculate gaps for customers with 2+ orders
df_multi = df_orders[df_orders['customer_id'].isin(customers_with_multiple.index)]
gaps_multi = df_multi.groupby('customer_id')['order_date'].diff().dt.days
```

```python
ans_q1 = gaps.dropna().median()
print(f"Giá trị trung vị tìm được là: {ans_q1}") 
```

```python
# Compare both approaches
print("Method 1 (original - all customers):")
print(f"  Median: {gaps.dropna().median()} days")
print(f"  Count of gaps: {gaps.dropna().count()}")

print("\nMethod 2 (explicit 2+ orders filter):")
median_multi = gaps_multi.dropna().median()
print(f"  Median: {median_multi} days")
print(f"  Count of gaps: {gaps_multi.dropna().count()}")

# They should be the same, but let's verify
print(f"\nAre they equal? {gaps.dropna().median() == median_multi}")
```

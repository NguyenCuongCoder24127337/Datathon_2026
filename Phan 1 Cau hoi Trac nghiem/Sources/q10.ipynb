```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df_payments = pd.read_csv('../data/payments.csv')
df_payments.head()
```

```python
installment_analysis = df_payments.groupby('installments').agg({
    'payment_value': 'sum',        # Tổng số tiền thanh toán của nhóm đó
    'order_id': 'nunique'          # Số lượng đơn hàng duy nhất của nhóm đó
}).reset_index()

# 3. Tính giá trị trung bình trên mỗi đơn hàng
installment_analysis['avg_value_per_order'] = (
    installment_analysis['payment_value'] / installment_analysis['order_id']
)
```

```python
best_plan = installment_analysis.loc[installment_analysis['avg_value_per_order'].idxmax()]

print(f"Kế hoạch trả góp tốt nhất là: {int(best_plan['installments'])} kỳ")
print(f"Giá trị trung bình: {best_plan['avg_value_per_order']:.2f}")
```

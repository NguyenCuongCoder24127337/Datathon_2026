```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df_products = pd.read_csv('../data/products.csv')
df_products.head(5)
```

```python
df_products.info()
```

```python
df_products['ave_gross_profit_margin'] = (df_products['price'] - df_products['cogs'])/df_products['price']
best_segment = df_products.groupby('segment')['ave_gross_profit_margin'].mean().idxmax()

print(f"Phân khúc có tỷ suất lợi nhuận cao nhất là: {best_segment}")
```

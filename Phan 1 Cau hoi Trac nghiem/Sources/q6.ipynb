```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_customers = pd.read_csv('../data/customers.csv')
df_orders = pd.read_csv('../data/orders.csv')
df_customers.info()
df_orders.info()
```

```python
df_merged = pd.merge(df_customers, df_orders, on='customer_id', how='left')
```

```python
# Lọc khách hàng có age_group khác null
df_merged_clean = df_merged[df_merged['age_group'].notna()]

# Nhóm tuổi nào có số đơn hàng trung bình cao nhất?
result = df_merged_clean.groupby('age_group').agg({
    'order_id': 'count',      # Tổng số đơn trong nhóm
    'customer_id': 'nunique'  # Số khách hàng duy nhất trong nhóm
}).reset_index()

result.columns = ['age_group', 'total_orders', 'num_customers']
result['avg_orders_per_customer'] = result['total_orders'] / result['num_customers']

print(result)
print(f"\nNhóm tuổi có đơn trung bình cao nhất: {result.loc[result['avg_orders_per_customer'].idxmax(), 'age_group']}")
```

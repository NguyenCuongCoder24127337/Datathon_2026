```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_geo = pd.read_csv('../data/geography.csv')
df_orders = pd.read_csv('../data/orders.csv')
df_orderitems = pd.read_csv('../data/order_items.csv')
df_orderitems.info()
df_orders.info()
```

```python
df_orderitems['revenue'] = df_orderitems['quantity'] * df_orderitems['unit_price']

df_merged = pd.merge(df_orders[['order_id', 'zip']], df_orderitems[['order_id', 'revenue']], on='order_id', how='inner')
```

```python
df_merged.info()
```

```python
df_merged = pd.merge(df_merged, df_geo[['zip', 'region']], on='zip', how='inner')
```

```python
final_revenue = df_merged.groupby('region')['revenue'].sum()
print(final_revenue.idxmax())
```

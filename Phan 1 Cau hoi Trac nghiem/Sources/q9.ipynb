```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df_order_items = pd.read_csv('../data/order_items.csv')
# df_order_items.head()
df_products = pd.read_csv('../data/products.csv')
# df_products.head()
df_returns = pd.read_csv('../data/returns.csv')
df_returns.head()
```

```python
# Merge để lấy thông tin size cho từng item đã bán
df_sales_size = pd.merge(df_order_items, df_products[['product_id', 'size']], on='product_id', how='inner')

# Đếm tổng số dòng bán ra cho mỗi size
sold_counts = df_sales_size.groupby('size').size()
```

```python
# Merge để biết món bị trả lại là size gì
df_returns_size = pd.merge(df_returns, df_products[['product_id', 'size']], on='product_id', how='inner')

# Đếm tổng số dòng bị trả lại cho mỗi size
return_counts = df_returns_size.groupby('size').size()
```

```python
# Chia tử số cho mẫu số
return_rate = (return_counts / sold_counts)

# Lấy ra size có tỷ lệ cao nhất
result = return_rate.idxmax()
print(f"Kích thước có tỷ lệ trả hàng cao nhất là: {result}")
print(return_rate.sort_values(ascending=False))
```
